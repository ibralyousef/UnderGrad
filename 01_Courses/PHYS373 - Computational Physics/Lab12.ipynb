```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
width = 2.0  # meters
height = 1.0  # meters
top_temp = 100.0  # degrees Celsius
bottom_temp = 50.0  # degrees Celsius
left_temp = 0.0  # degrees Celsius
right_temp = 75.0  # degrees Celsius
num_points_x = 5
num_points_y = 6
initial_guess = 25.0  # degrees Celsius
convergence_criterion = 0.01  # degrees Celsius

# Calculate grid spacing
delta_x = width / (num_points_x - 1)
delta_y = height / (num_points_y - 1)

# Create grid
x = np.linspace(0, width, num_points_x)
y = np.linspace(0, height, num_points_y)
X, Y = np.meshgrid(x, y)

# Initialize temperature grid
T = np.full((num_points_y, num_points_x), initial_guess)

# Define boundary conditions
T[0, :] = top_temp
T[-1, :] = bottom_temp
T[:, 0] = left_temp
T[:, -1] = right_temp

# Main iteration loop
converged = False
while not converged:
    T_new = np.copy(T)  # Create a copy of the temperature grid

    # Update interior points
    for i in range(1, num_points_y - 1):
        for j in range(1, num_points_x - 1):
            T_new[i, j] = (T[i-1, j] + T[i+1, j] + T[i, j-1] + T[i, j+1]) / 4

    # Check convergence
    max_diff = np.max(np.abs(T_new - T))
    if max_diff <= convergence_criterion:
        converged = True

    T = T_new  # Update the temperature grid

# Create heat map plot
plt.figure()
plt.pcolormesh(X, Y, T, cmap='hot', shading='auto')
plt.colorbar(label='Temperature (Â°C)')
plt.xlabel('X-axis (m)')
plt.ylabel('Y-axis (m)')
plt.title('Steady-State Temperature Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
width = 2.0  # meters
height = 1.0  # meters
top_temp = 100.0  # degrees Celsius
bottom_temp = 50.0  # degrees Celsius
left_temp = 0.0  # degrees Celsius
right_temp = 75.0  # degrees Celsius
num_points_x = 50
num_points_y = 60
initial_guess = 25.0  # degrees Celsius
convergence_criterion = 0.001  # degrees Celsius

# Calculate grid spacing
delta_x = width / (num_points_x - 1)
delta_y = height / (num_points_y - 1)

# Create grid
x = np.linspace(0, width, num_points_x)
y = np.linspace(0, height, num_points_y)
X, Y = np.meshgrid(x, y)

# Initialize temperature grid
T = np.full((num_points_y, num_points_x), initial_guess)

# Define boundary conditions
T[0, :] = top_temp
T[-1, :] = bottom_temp
T[:, 0] = left_temp
T[:, -1] = right_temp

# Main iteration loop
converged = False
while not converged:
    T_new = np.copy(T)  # Create a copy of the temperature grid

    # Update interior points
    for i in range(1, num_points_y - 1):
        for j in range(1, num_points_x - 1):
            T_new[i, j] = (T[i-1, j] + T[i+1, j] + T[i, j-1] + T[i, j+1]) / 4

    # Check convergence
    max_diff = np.max(np.abs(T_new - T))
    if max_diff <= convergence_criterion:
        converged = True

    T = T_new  # Update the temperature grid

# Create 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, T, cmap='hot')
ax.set_zlabel('Temperature (Â°C)')
ax.set_title('Steady-State Temperature Distribution')
# add colorbar
m = plt.cm.ScalarMappable(cmap='hot')
m.set_array(T)
plt.colorbar(m, ax=ax)
plt.gca().invert_yaxis()
plt.show()
```

1. The top region has the highest temperature, the left one has the lowest.
2. The temperature flow from the hotter side to the cooler one, as we can see in the 3D plot
3. The boundaries supply a constant heat flow, therefore after long time the solution will be the same as the 3D plot
